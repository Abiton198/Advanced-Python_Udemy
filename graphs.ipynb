```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
names = ['Tyris' , 'Travis','Cherise','Michelle', 'Natty', 'Nicolle']
marks =[76,66,74,77,88,92]
```

```python
# plt.plot(names,marks)
plt.bar(names, marks, color = 'red', width = 0.4, edgecolor = 'green', linestyle = '-', linewidth = 3, yerr=3, ecolor = 'yellow', alpha = 0.5)
plt.title('Class Results')
plt.ylabel('Total Class Marks')
plt.xlabel('Names')
plt.show()
#alpha = shade color, align = alignment, yerr = lines in bar charts
```

```python
hrs = df.iloc[0:10 ,0]
att = df.iloc[0:10 ,1]
x_axis = plt.plot(hrs, color = 'red')
y_axis = plt.plot(att, color = 'blue')
plt.title('Comparison Hours and Attendance')
plt.xlabel('Hours')
plt.ylabel('Attendance')
plt.show()
```
