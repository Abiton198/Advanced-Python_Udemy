```python
# Node class represents one element in the linked list
class Node:
    def __init__(self, data):
        # Store the actual data inside the node
        self.data = data

        # Pointer to the next node (initially nothing)
        self.next = None

        # Pointer to the previous node (used in doubly linked list)
        # NOTE: You are using prev in add(), so we define it here
        self.prev = None


# LinkedList class manages the whole list
class LinkedList:
    def __init__(self):
        # Head is the first node in the list
        # Initially the list is empty
        self.head = None


    # Method to add a new node at the beginning
    def add(self, value):

        # Step 1: Create a new node with the given value
        a = Node(value)

        # Step 2: Make the new node point to the current head
        # (new node → old head)
        a.next = self.head

        # Step 3: If list is not empty,
        # make old head point back to new node
        if self.head != None:
            self.head.prev = a

        # Step 4: Move head to the new node
        # (new node becomes the first node)
        self.head = a


    # Method to print the list
    def printList(self):

        # Start from the head
        current = self.head

        # Traverse until we reach the end (None)
        while current is not None:

            # Print current node's data
            print(current.data)

            # Move to the next node
            current = current.next


# Create linked list object
x = LinkedList()

# Add a node
x.add('ABITON')
x.add('Udemy')
x.add('Michelle')

# Print the list
x.printList()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    # Add at beginning
    def add(self, value):
        a = Node(value)
        a.next = self.head
        self.head = a

    # Insert after a given node
    def insertion(self, node, value):

        # Safety check
        if node is None:
            print("The given node cannot be None")
            return

        new = Node(value)
        new.next = node.next
        node.next = new

    def appending(self,value):
        new = Node(value)
        new.next = None
        if self.head != None:
            new.prev = None
            self.head = new
            return
        pointer = self.head
        while pointer.next != None:
            pointer = pointer.next
        pointer.next = new
        new.prev = pointer
        return


    def printList(self):
        current = self.head
        while current is not None:
            print(current.data)
            current = current.next


# Create list
x = LinkedList()

# Add some nodes first
x.add(10)
x.add(20)
x.add(30)

# Current list:
# 30 → 20 → 10

# Insert 12 after second node (20)
x.insertion(x.head.next, 12)
x.appending(x.head.next)
x.printList()
```
