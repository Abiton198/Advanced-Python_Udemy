```python
pattern = r'Abiton'
sequence = 'Abitn'
x = re.match(pattern, sequence)
if x:
    print('Matched')
else:
    print('Not Matched')
print(x)
```

```python
#grouping
pattern = 'Abiton Padera is AN nlp engineer'
x = re.match(r'(Abiton) (Padera) (is)',pattern)
if x:
    print(x.group())
    print(x.group(1))
    print(x.group(1,2))
    print(x.group(3))
```

```python
#matching
x = re.search(r'Abiton\\Padera', 'Abiton\Padera').group()
print(x)
```

```python
#\s means a space between words
y = re.search(r'Abiton\sPadera', 'Abiton Padera').group()
print(y)
```

```python
#match search ?<=
x = re.search(r'(?<=abc)', 'abcxyz').group()
if x:
    print('Matched')
else:
    print('Not Matched')
```

```python
pattern = '^d'
sequence = 'abcdef'
x = re.match(pattern, sequence)
if x:
    print('Matched')
else:
    print('Not Matched')
```

```python
#greedy & non greedy match qualifier
sequence = '<h1>Title</h1>'
x = re.match(r'<.*>',sequence).group() #greedy
y = re.match(r'<.*?>',sequence).group() #non-greedy ?
print(x)
print(y)
if x:
    print('Matched')
else:
    print('Not Matched')
```

```python
#case sensitive match
#re.I / re.s / re.X / re.M / re.L / re.S / re.U
pattern = 'Abiton \n Padera'
sequence = 'abiton \n padera'
x = re.match(pattern, sequence)
if x:
    print('Matched')
x = re.match(pattern, sequence, re.M)
print(x)
```

```python
#VALIDATION
passwords = ['abi123','erf123','asd145','zxc','vbn234','mnb']
matching_sequence = r'[0-9]'

for x in passwords:
    r = re.search(matching_sequence, x)
    if r:
       print(x, ' is valid password')
    else:
        print(x, ' is not a valid password')
```
