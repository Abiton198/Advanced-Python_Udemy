```
#convert dict into dataframe

eduplanet = {'name':['Xola','Zola','Asakhile','Anganani','Aphiwe','Andiswa'],
             'marks':[36,56,78,23,67,55]}

cohan = {'name':['Xola','Zola','Asakhile','Anganani','Aphiwe','Aaroosa','Tyris', 'Travis', 'Luke'],
             'sports':['basketball','golf','soccer','fishing','rugby','cricket','basketball','golf','soccer']}

edu = pd.DataFrame(eduplanet)
coh = pd.DataFrame(cohan)
```

```

coh.head()
```

```
edu.head()
```

```
merged_df = pd.merge(edu,coh,on='name', how='right')
merged_df
```

```
merged_df['Rank'] = merged_df['marks'].rank(ascending=1, method='min')
merged_df.head()
```

```
merged_df.sort_values(['Rank','marks'], ascending=[0,1], inplace=True)
```

```
merged_df
```

```
def comment(x):
    if x <= 50:
        return "Below Average"
    else:
        return "Above Average"
```

```
#Add comment by executing function
merged_df['comment'] = merged_df['marks'].apply(comment)
merged_df.head()
```

```
#write to excel
writer = pd.ExcelWriter(r'C:\Users\michelle\Desktop\students.xlsx')
merged_df.to_excel(writer, index=False, header=False)
writer._save()
writer.close()
```

```
#isin()
boolen_series = merged_df['sports'].isin(['basketball','soccer'])
boolen_series.head()
new_df = merged_df[boolen_series]
```
